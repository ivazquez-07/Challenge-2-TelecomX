#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import requests
import json

url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'

# MÃ©todo fÃ¡cil
# datos_json = pd.read_json(url)
# datos_json.head()

# MÃ©todo mÃ¡s dificil usando API
get_api = requests.get(url)
datos_api_json = json.loads(get_api.text)
datos_json = pd.DataFrame(datos_api_json)
datos_json.head()
```

Esta informaciÃ³n presenta informaciÃ³n anidada, debemos normalizarlo

```python
datos_customer = pd.json_normalize(datos_json['customer'])
datos_customer.head(2)
```

```python
datos_phone = pd.json_normalize(datos_json['phone'])
datos_phone.head(2)
```

```python
datos_internet = pd.json_normalize(datos_json['internet'])
datos_internet.head(2)
```

```python
datos_account = pd.json_normalize(datos_json['account'])
datos_account.head(2)
```

#ðŸ”§ TransformaciÃ³n

```python
df_sin_anidados = datos_json.drop(columns=['customer', 'phone', 'internet', 'account'])
df_combinados = pd.concat([df_sin_anidados, datos_customer, datos_phone, datos_internet, datos_account], axis=1)
display(df_combinados.head())
```
